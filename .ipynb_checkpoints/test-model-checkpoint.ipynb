```python
import numpy as np
import pandas as pd
import json
```

```python
import os
os.getcwd()
```

```python
data = open('training_data/class__software_engineering/class__software_engineering.json')
data = json.load(data)
```

```python
labels = np.array(list(data.keys()))
```

```python
labels
```

```python
arr, names = [], []
```

```python
for label in labels:
    for encoding in data[label]:
        temp = encoding
        names.append(label)
        arr.append(temp)
```

```python
data = pd.DataFrame(arr)
```

```python
data
```

```python
data['labels'] = names
```
